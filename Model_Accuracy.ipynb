```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
AlexNet = pd.read_csv("log1.csv",sep =';')
AlexNet.head()
```

```python
plt.plot(AlexNet['epoch'], AlexNet['loss'], 'g', label='Training loss')
plt.plot(AlexNet['epoch'], AlexNet['val_loss'], 'b', label='validation loss')

plt.title('Training and Validation loss - AlexNet')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig("AlexNet_loss.png")
```

```python
plt.plot(AlexNet['epoch'], AlexNet['accuracy'], color = 'orange', label='Training acc')
plt.plot(AlexNet['epoch'], AlexNet['val_accuracy'], 'r', label='validation acc')
plt.title('Training and Validation acc - AlexNet')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig("AlexNet_acc.png")
```
